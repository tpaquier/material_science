```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
%run rbf.ipynb
np.random.seed(167176)
noise = np.random.uniform(low=-1.,high=1,size=100)
red_cluster = np.random.multivariate_normal(mean=np.array((0,1)),cov=np.eye(2)*0.25,size=100)
```

```python
X = np.linspace(start=-3.5,stop=3.5,num=100)
Y = 0.5*X**2-2+noise
```

```python
plt.plot(red_cluster[:,0],red_cluster[:,1], 'ro', label='red cluster')
plt.plot(X, Y,'bo',label='blue cluster')
plt.legend()
```

```python
W = np.zeros((200,200))
```

```python
blue_data = np.zeros((100,2))
blue_data[:,0] = X
blue_data[:,1] = Y
whole_data = np.vstack((red_cluster,blue_data))
```

```python
for i in range(200):
    for k in range(i,200):
        W[i,k] = rbf(x=whole_data[i,:],y=whole_data[k,:],l_squared=2)
        W[k,i] = W[i,k]
    W[i,i] = 1
```

```python
diag_mat = np.zeros((200,200))
```

```python
for i in range(200):
    diag_mat[i,i] = np.sum(W[:,i])
```

```python
L = diag_mat - W
```

```python
eigen = np.linalg.eig(L)
```

```python
eigenvalues, eigenvectors = eigen
```

```python
for i in range(eigenvalues.shape[0]):
    if i<1e-10:
        eigenvalues[i] = 0

sorted_eigenvalues = np.argsort(eigenvalues)
```

```python
n_cluster = 3
U = np.zeros((200,n_cluster))
for i in range(n_cluster):
    U[:,i] = eigenvectors[:,sorted_eigenvalues[i]]
```

```python
cluster_fit = KMeans(n_clusters=n_cluster).fit(U)
```

```python
labels = cluster_fit.predict(U)
```

```python
df_points = pd.DataFrame(whole_data)
```

```python
df_points['label'] = labels
df_points.columns = ['x1','x2','label']
```

```python
colors = {0: 'orange', 1: 'purple',2:'blue',3:'red',4:'yellow'}
color_list = [colors[group] for group in df_points['label']]
```
