```python
import numpy as np
from sklearn.cluster import KMeans, SpectralClustering
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(18768787)
blue = np.random.multivariate_normal(mean=np.array((1,1)), cov=np.array(([2.,-0.5],[-0.5,0.15])),size=100)
red = np.random.multivariate_normal(mean=np.array((2,2)), cov=np.array(([2.,-0.5],[-0.5,0.15])),size=100)
```

```python
plt.plot(blue[:,0],blue[:,1],'bo')
plt.plot(red[:,0], red[:,1], 'ro')
```

```python
whole_data = np.vstack((red,blue))
```

```python
kmeans = KMeans(n_clusters=2).fit_predict(whole_data)
```

```python
kmeans_df = pd.DataFrame(whole_data)
kmeans_df.columns = ['x1','x2']
kmeans_df['label'] = kmeans
```

```python
colors = {0: 'blue', 1: 'red'}
color_list = [colors[group] for group in kmeans_df['label']]
```

```python
ax = kmeans_df.plot.scatter('x1','x2', c=color_list)
```

```python
spectral = SpectralClustering(n_clusters=2,n_components=2).fit_predict(whole_data)
```

```python
spectral_df = pd.DataFrame(whole_data)
spectral_df.columns = ['x1','x2']
spectral_df['label'] = spectral
```

```python
colors_bis = {0: 'blue', 1: 'red'}
color_list_bis = [colors_bis[group] for group in spectral_df['label']]
```

```python
ax = spectral_df.plot.scatter('x1','x2', c=color_list_bis)
```
